```python
import matplotlib.pyplot as plt
```

# Line Chart

```python
month = ['Jan','Feb','Mar','Apr','May']
coconut_yield = [2000,2500,2200,2300,2600]
plt.plot(month,coconut_yield,marker="*")
plt.title("Monthly Coconut yield data")
plt.xlabel("Month")
plt.ylabel("Yield")
plt.show()
```

# Bar Chart

```python
dept = ["BCA","BSc","BA.Tamil"]
students_average = [92,80,91]
plt.bar(dept,students_average)
plt.title("Average marks of the student across dept")
plt.xlabel("Department")
plt.ylabel("Students Average mark")
plt.show()
```

# Pie Chart

```python
share_holding = ["Promoter holding","Foreign Investerss","Domestic investers","Public investers","Mutual funds"]
percentage = [42.56,17.88,15.08,24.14,10.01]
plt.pie(percentage,labels=share_holding,autopct='%1.1f')
plt.title("Tata Share holdings")
plt.show()
```

# Subplot

```python
months = ["jan","feb","mar","apr"]
 
sales = [100,200,150,120]
profit = [20,25,30,35]
 
plt.figure(figsize=(10,4))
 
plt.subplot(1,2,1)
plt.plot(months,sales)
plt.title("monthly sales")
 
plt.subplot(1,2,2)
plt.plot(months,profit)
plt.title("profit")
```

# Histogram

```python
# Perfect for seeing the distribution or spread of a single column
# ---------------------------------------------------------
plt.figure(figsize=(8, 5))

# 'bins' define how many groups/bars the data is split into
# 'edgecolor' makes the bars visually distinct
plt.hist(df['Age'], bins=5, edgecolor='black', color='skyblue')

# Adding labels and title
plt.title('Distribution of Employee Ages')
plt.xlabel('Age Group')
plt.ylabel('Frequency (Number of Employees)')

# Display the plot
plt.show()
```

# Practise Project

```python
import pandas as pd
 
data = {

    "Month":["Jan","Feb","Mar","Apr","May","Jun",

             "Jul","Aug","Sep","Oct","Nov","Dec"],
 
    "Sales":[50000,55000,60000,65000,70000,75000,

             80000,85000,90000,95000,100000,110000],
 
    "Profit":[10000,12000,15000,16000,18000,20000,

              22000,24000,26000,28000,30000,35000]

}
 
df = pd.DataFrame(data)
 
print(df)
 
```

```python
plt.plot(df['Month'],df['Profit'],marker="o")
plt.title("Monthly profit")
plt.xlabel("Month")
plt.ylabel("Profit")
plt.show()
```

```python
df['Expense'] =df['Sales']-df['Profit']
print(df)
plt.bar(df['Month'],df['Expense'])
plt.title("Monthly Expense")
plt.xlabel("Month")
plt.ylabel("Expense")
plt.show()
```

```python
data1 = df['Profit'].sum()
data2 = df['Profit'].sum()
data3 = df['Profit'].sum() #using same data
yearly_profit = [data1,data2,data3]

plt.pie(yearly_profit,labels=yearly_profit,autopct='%1.1f')
plt.title("Yearly Profit")
plt.show()
```
