```python
import pandas as pd
import numpy as np
```

```python
# Step 1: Load the data
try:
    df = pd.read_csv('messy_dataset.csv')
except FileNotFoundError:
    print("File Not Found")
except Exception:
    print("Error Occured")
else:
    print("File Successfully Loaded")
```

```python
# Step 2: Rename messy columns
print("Before Renaming")
print(df,"\n")
df.rename(columns={'cust_nm':'Name', 'sales_amt':'Sales','age':'Age','region':'Region'}, inplace=True)
print("After Renaming ...")
print(df)
```

```python
# Step 3: Fill missing Sales with mean

#df['Sales'].fillna(df['Sales'].mean(), inplace=True)

# Step 3: (Fixed for Pandas 3.0)

sales_mean = df['Sales'].mean()
df['Sales'] = df['Sales'].fillna(sales_mean)

print("After updating sales")
print(df)
```

```python
# Step 4: Drop rows with missing Age
df.dropna(subset=['Age'], inplace=True)

print("After removing NaN from Age Column")
print(df)
```

```python
# Step 5: Remove duplicates
df.drop_duplicates(inplace=True)
print("After removing duplicates from the dataframe")
print(df)
```

```python
print('Cleaned Data - shape : ', df.shape)
print(df)
```
