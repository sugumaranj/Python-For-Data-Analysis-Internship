## Practical 1 : Attendance Calculator

```python
def attendance_percentage(attended, total):
  percentage = (attended/total) * 100
  return round(percentage,2)

def check_eligibility(percentage):
  if percentage >= 75:
    return "Eligible"
  else:
    return "Not Eligible"

try:
  n = int(input("Enter number of students: "))
  for i in range(1,n+1):
    total = int(input("Enter total number of classes: "))
    attended = int(input("Enter number of classes attended: "))
    percentage = attendance_percentage(attended, total)
    status = check_eligibility(percentage)

    print(f"Attendance percentage: {percentage}%")
    print(f"Status: {status}")
    print("\n")

except ValueError:
  print("Invalid input. Please enter valid numbers.")

except Exception:
  print("An error occurred. Please try again.")
else:
  print("Program fully executed without any errors.")
finally:
  print("Program execution completed")
```

```python
import csv

def read_students(student_file):
  with open(student_file, 'r') as file:
    reader = csv.DictReader(file)
    students = []
    for row in reader:
      students.append(row)
  return students

with open("students_record.csv","w",newline="") as csv_file:
    fieldnames=['Name','Department','Mark']

    writer = csv.DictWriter(csv_file,fieldnames=fieldnames)
    writer.writeheader()
    #for to insert multiple row at once - writerows() function
    writer.writerow({'Name': 'Alice', 'Department': 'Maths', 'Mark': 90})
    writer.writerow({'Name': 'Bob', 'Department': 'Physics', 'Mark': 85})
    writer.writerow({'Name': 'Charlie', 'Department': 'Chemistry', 'Mark': 78})
    writer.writerow({'Name': 'David', 'Department': 'Biology', 'Mark': 92})


students=read_students("students_record.csv")


print(f"{'Name':<12}{'Department':<12}{'Mark':>5}")
print('-' * 30)

for student in students:
  name = student.get('Name',"N/A")
  dept = student.get('Department',"N/A")
  mark = student.get('Mark',"N/A")
  print(f"{name:<12}{dept:<12}{mark:>5}")

print(f"\nTotal students : {len(students)}")
```
