```python
import numpy as np
```

## Practise 1 : Student Score analysis

```python
#scores = [78,85,92,67,88]
#total = sum(scores)
#average = total / len(scores)
#highest = max(scores)
#lowest = min(scores)


scores = np.array([78,85,92,67,88])
total = scores.sum()
average = scores.mean()
highest = scores.max()
lowest = scores.min()

print("Student Scores : ",scores)
print("Total Marks : ",total)
print("Average Marks : ",average)
print("Highest Mark : ",highest)
print("Lowest Mark : ",lowest)
```

## Practise 2 : Sales Array Calculations

```python
sales = [12000,15000,18000,14000,20000]

#total_sales = sum(sales)
#average_sales = total_sales / len(sales)
#highest_sale = max(sales)
#lowest_sale = min(sales)

sales = np.array(sales)
total_sales = sales.sum()
average_sales = sales.mean()
highest_sale = sales.max()
lowest_sale = sales.min()

print("Monthly Sales : ",sales)
print("Total Sales : ",total_sales)
print("Average Sales : ",average_sales)
print("Highest Sale : ",highest_sale)
print("Lowest Sale : ",lowest_sale)
```
