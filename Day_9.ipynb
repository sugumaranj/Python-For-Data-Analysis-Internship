# Student Performance Dashboard Program

```python
import pandas as pd
import matplotlib.pyplot as plt
 
data = {
    'Student':['A','B','C','D','E','F','G','H'],
    'Marks':[85,72,90,65,78,88,60,95],
    'Attendance':[95,85,98,75,88,92,70,99],
    'Study_Hours':[8,6,9,4,5,8,3,10]
}
 
df = pd.DataFrame(data)
```

# Histogram

```python
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
plt.hist(df['Marks'], bins=5)
plt.title("Marks Distribution")
```

# Scatter Plot

```python
plt.subplot(2,2,2)
plt.scatter(df['Study_Hours'], df['Marks'])
plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
```

# Bar Chart

```python
plt.subplot(2,2,3)
plt.bar(df['Student'], df['Marks'])
plt.title("Student Marks")
```

# Pie Chart

```python
plt.subplot(2,2,4)
plt.pie(df['Attendance'],
        labels=df['Student'],
        autopct='%1.1f%%')
plt.title("Attendance Share")

plt.tight_layout()
plt.show()
```
