## Practise Project 1 - Student Marks Calculator

```python
marks = [50,60,70,80,100]
total = sum(marks)
average = total / len(marks)

if (average >= 90 ):
  grade = "S (Outstanding)"
elif (average >= 75):
  grade = "A (Distinction)"
else:
  grade = "B or below"

print("\t--------- Result Card ---------")
print(f"Total Marks: {total} / 500")
print(f"Average Marks: {average:.2f}%")
print(f"Grade: {grade}")
```

## Practise Project 2 - Expense Tracker

```python
budget = 25000
expense = {"Food":2500,"Rent":6000,"Subscriptions":5000} #category : value
total = 0

print("------------Expenses------------")
for category,amount in expense.items():
  print(f"{category}: {amount}")
  total += amount

print("--------------------------------")
print("Total Expense : ",total)
print("--------------------------------")


if total > budget:
  print("Over Budget")
else:
  print("Under Budget")
```

## Practise Project 3 - Grading System

```python
def get_grade(mark):
  if(mark>=90):
    return 'S'
  elif(mark>=75):
    return 'A'
  else:
    return 'F'

marks = {"Prasanna":98,"Siva":67,"Jack":76}
fail_count = 0
pass_count = 0

for name,mark in marks.items():
    grade = get_grade(mark)
    print(f"{name}: {mark} ({grade})")
    if(grade == 'F'):
      fail_count += 1
    else:
      pass_count += 1

print("--------------------------------")
print(f"Total Passed: {pass_count}")
print(f"Total Failed: {fail_count}")
```
