<h2>Employee dataset exploration</h2>

```python
import pandas as pd
employees = {
    "emp_id":[1,2,3,4,5],
    "name":["Arun","Priya","Rahul","Divya","Kumar"],
    "department":["IT","HR","IT","Finance","HR"],
    "salary":[45000,40000,50000,55000,42000],
    "experience":[2,3,5,6,4]
}
```

```python
df = pd.DataFrame(employees)
print("Employee Dataset")
df
```

```python
print("Average Salary")
print(df["salary"].mean())
```

```python
print("Highest Salary")
print(df["salary"].max())
```

```python
print("Employees in HR")
print(df[df["department"]=="HR"])
```

```python
print("Employees with Experience >3")
print(df[df["experience"]>3])
```

```python
print("Employee Count by Department")
print(df["department"].value_counts())
```
